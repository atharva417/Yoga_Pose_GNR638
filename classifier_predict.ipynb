```
model = models.load_model('/content/drive/My Drive/yoga_set/yoga_ges_80_85.h5')
```

```
import numpy as np
import os
import glob
from keras.preprocessing import image
classess = []
filenames = [img for img in glob.glob("/content/drive/MyDrive/yoga_set/test_data/*.png")]
filenames.sort()

for img in filenames:
  img = image.load_img(img, target_size=(256,256))
  img = image.img_to_array(img)/255
  img = np.expand_dims(img, axis=0)
  result = model.predict(img)
  clas = np.argmax(result)
  classess.append(clas)
```
